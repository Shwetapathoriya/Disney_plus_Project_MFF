```python
#question 1
#"1.Is there a relationship between the release year and the rating ?";
```

```python


# Load the dataset
file_path = 'disney_plus_titles.csv'
disney_data = pd.read_csv(file_path)

# Group by release year and rating, and count occurrences
rating_distribution = disney_data.groupby('release_year')['rating'].value_counts().unstack().fillna(0)

# Sum the counts of ratings across all years
rating_sum = rating_distribution.sum(axis=0)

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(rating_sum, labels=rating_sum.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings Across All Release Years')
plt.show()
```

```python

# Load the dataset
file_path = 'disney_plus_titles.csv'
disney_data = pd.read_csv(file_path)

# Create the count plot
plt.figure(figsize=(14, 7))
ax = sns.countplot(data=disney_data, x='rating', hue='release_year')
plt.title('Distribution of Content Ratings by Release Year on Disney+')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Release Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Question 3
#3.How many movies and TV shows are there on Disney+?
```

```python

# Load the dataset
file_path = 'disney_plus_titles.csv'
disney_data = pd.read_csv(file_path)

# Filter for movies and TV shows
movies_data = disney_data[disney_data['type'] == 'Movie']
tv_shows_data = disney_data[disney_data['type'] == 'TV Show']

# Count the number of movies and TV shows
num_movies = movies_data.shape[0]
num_tv_shows = tv_shows_data.shape[0]

# Plot the counts
plt.figure(figsize=(8, 6))
sns.barplot(x=['Movies', 'TV Shows'], y=[num_movies, num_tv_shows])
plt.title('Number of Movies and TV Shows on Disney+')
plt.ylabel('Count')
plt.show()
```

```python
# question 4
# "4.What is the average number of genres per title on Disney+?"
```

```python

# Load the dataset
file_path = 'disney_plus_titles.csv'
disney_data = pd.read_csv(file_path)

# Extract genres from 'listed_in', splitting by comma and stacking
genres_series = disney_data['listed_in'].str.split(', ', expand=True).stack()
genres_df = genres_series.reset_index(level=1, drop=True)
genres_df.name = 'genre'

# Create a DataFrame with genres and their associated release years
genres_df = pd.concat([disney_data[['release_year']], genres_df], axis=1)
genres_df = genres_df.dropna(subset=['release_year'])

# Plot the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='genre', y='release_year', data=genres_df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Distribution of Release Years by Genre on Disney+')
plt.xlabel('Genre')
plt.ylabel('Release Year')
plt.show()
```

```python
  #Question.6
  #"6.What are the most frequently occurring keywords in the descriptions of Disney+ content?"
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK data files (only need to do this once)
nltk.download('punkt')
nltk.download('stopwords')



# Combine all descriptions into one large text
text = ' '.join(data['description'].dropna().tolist())

# Tokenize the text
tokens = word_tokenize(text.lower())

# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
words = [word for word in tokens if word.isalnum() and word not in stop_words]

# Count word frequencies
freq_dist = nltk.FreqDist(words)

# Create a DataFrame from the frequency distribution
freq_df = pd.DataFrame(freq_dist.items(), columns=['Word', 'Frequency'])
freq_df = freq_df.sort_values(by='Frequency', ascending=False)

# Print top 10 most frequent words
print(freq_df.head(10))

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequently Occurring Keywords in Disney+ Descriptions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame loading (replace this with your actual data loading)
# df = pd.read_csv('your_data.csv')

# Drop rows where 'director' is missing
data = data.dropna(subset=['director'])

# Split multiple directors and normalize the DataFrame
directors = data['director'].str.split(',', expand=True).stack()
directors_data= pd.DataFrame(directors, columns=['director'])
directors_data['director'] = directors_data['director'].str.strip()  # Remove extra spaces

# Count the number of titles for each director
director_counts = directors_data['director'].value_counts()

# Bar Plot the results
plt.figure(figsize=(12, 8))
director_counts.head(20).plot(kind='bar', color='teal')
plt.title('Top Directors with the Most Titles on Disney+')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
